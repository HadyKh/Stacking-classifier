```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, StackingClassifier
from sklearn.base import clone
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
```

```python
rs = 0
x_circles, y_circles = make_circles(300, noise = 0.1, random_state = rs)
circle_x_train, circle_x_test, circle_y_train, circle_y_test = train_test_split(x_circles, y_circles,
                                                                                test_size = 0.2, random_state = rs)
```

```python
def plotting(x, y, model, title):
    cls =len(set(y))
    X, Y = np.array(x), np.array(y) #convert to numpy
    plt.figure(figsize=(7,5), dpi=250)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.2),
                         np.arange(y_min, y_max, 0.2))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.YlGnBu)

    plt.xlabel('X1')
    plt.ylabel('X2')
    
    plot_colors = ['orange','green','purple']
    # Plot the training points
    for i, color in zip(range(cls), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color,
                    cmap=plt.cm.YlGnBu, edgecolor='black', s=10)

    plt.title(title)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.axis("tight")
    plt.show()
```

# Base estimators

```python
#Define estimators
estimators = [DecisionTreeClassifier(criterion = 'entropy', random_state = 0),
              AdaBoostClassifier(n_estimators = 50, learning_rate = 1, random_state=0),
             RandomForestClassifier(n_estimators = 5, random_state = 0)]
estimators = {estimator.__class__.__name__: estimator for estimator in estimators}

print(f'Baselines Weighted accuracy Score:')

#gets the unfitted models to fit
for estName, est in estimators.items():
    est = clone(est).fit(circle_x_train, circle_y_train)
    predY = est.predict(circle_x_test)
    accuracy = accuracy_score(circle_y_test, predY)
    print(f'{estName}: {accuracy}')
```

# stacking with Logistic regression aggregator

```python
#logistic regression aggregrator
print(f'Stacking Weighted accuracy Score (Aggregator LR)')
stacking = StackingClassifier([(estName, clone(est)) for estName, est in estimators.items()],
                              LogisticRegression(random_state = 0), n_jobs = -1)
stacking.fit(circle_x_train, circle_y_train)
predY = stacking.predict(circle_x_test)
accuracy = accuracy_score(circle_y_test, predY)
print(f'Stacking (Aggregator LR): {accuracy}')
plotting(circle_x_test, circle_y_test, stacking, 'Logistic Regression with Stacking')
```

# stacking with GuassianNaiveBayes aggregator

```python
#Naive Bayes aggregrator
print(f'Stacking Weighted accuracy Score (Aggregator LR)')
stacking = StackingClassifier([(estName, clone(est)) for estName, est in estimators.items()],
                              GaussianNB(), n_jobs = -1)
stacking.fit(circle_x_train, circle_y_train)
predY = stacking.predict(circle_x_test)
accuracy = accuracy_score(circle_y_test, predY)
print(f'Stacking (Aggregator NB): {accuracy}')
plotting(circle_x_test, circle_y_test, stacking, 'Naive Bayes with Stacking')
```
